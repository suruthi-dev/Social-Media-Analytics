# Analysing Peterson Graph

The Peterson graph is a simple undirected graph named after the mathematician Julius Petersen.

Nodes: The Peterson graph has 10 nodes, typically denoted by the numbers 0 to 9.

Edges: There are 15 edges in the Peterson graph, connecting the vertices in a specific pattern.

```python
petersen = nx.petersen_graph()
```

```python
nx.draw(petersen,with_labels=True)
plt.show()
```

## Adjacency view

 adjacency list of a graph

```python
print(petersen.adj)
```

```python
print(petersen.degree)
```

### Degree of a vertex

```python
for i in petersen:
  print(f"{i} - Degree {len(petersen[i])}")
```

# Weighted Graph

 weighted graph represents a graph with weighted edges

```python
mo_fy = nx.Graph()
```

```python
# 10 - direct connection
# 9 - partnership
# 8 - step
```

```python
family = [
    ("Jay","Gloria",9),("Gloria","Manny",10),("Jay","Manny",8),
    ("Jay","Claire",10),("Jay","Mitchell",10),
    ("Claire","Phil",9),("Mitchell","Cam",9),
    ("Cam","Lily",10)
]
```

```python
mo_fy.add_weighted_edges_from(family)
```

```python
nx.draw(mo_fy,with_labels=True)
```

# Plotting the weights

```python
pos = nx.spring_layout(mo_fy)
nx.draw(mo_fy,pos,with_labels=True)

edge_weight = nx.get_edge_attributes(mo_fy,"weight")
nx.draw_networkx_edge_labels(mo_fy,pos,edge_labels = edge_weight)
plt.show()
```

# **Degree centrality**

There are two kinds of graphs  directed and non-directed.

The degree of a node in a non-directed graph is defined as the number of links a node has with other nodes.

 In a directed graph, the degree is subdivided into **in and out degrees**

 In-degree refers to the incoming links to the node while out-degree is the number of nodes directed at other nodes from a particular node.

**The higher the degree, the more crucial it becomes in the graph**
